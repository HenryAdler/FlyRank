## 1. My lane (or freestyle) and why

**Selected Lane:** Lane 2 — Refresh / Content Opportunity Scoring

I have chosen Lane 2 because it directly addresses the operational bottleneck faced by content and SEO teams: limited human review capacity. Instead of manually auditing thousands of pages or blindly updating content based on gut feeling, this project focuses on building an evidence-based decision-support system to rank content assets by their urgency and opportunity for optimization, ensuring editorial effort is channeled where it yields the highest probability of recovery.

## 2. The question: decision, action, cost of a wrong call

- **Research Question:** Which visible pages present observable signals of decay or optimization opportunities and should be prioritized for review first?
- **Unit of Analysis (Grain):** A unique pseudonymized content item (`content_id` / `content_hash_id`) per client.
- **The Decision & Action:** The content team decides to pause new content creation to execute an optimization action (e.g., rewriting meta tags, expanding thin text, or adjusting search intent match) on a specific page flagged by the queue.
- **Cost of a Wrong Call (False Positive):** Recommending a healthy page for a refresh results in wasted editorial hours and opportunity cost, potentially disrupting a page that was already performing well.
- **Cost of an Omission (False Negative):** Failing to flag a decaying page leads to a continuous, unmitigated loss of organic traffic, conversions, and market share to competitors.
- **Why ML/Data helps:** The sheer scale of content inventories makes manual auditing impossible. Machine learning can simultaneously weigh multiple historical indicators (impressions, clicks, age, positions, and drift) to output a dynamic, calibrated priority score, going far beyond rigid, static SQL rules.

## 3. Quick look at the data (2-3 real numbers)

Below, we load the starter dataset to extract key baseline numbers that justify dedicating the next 7 weeks to this lane.

```python
import pandas as pd
import numpy as np

# Load the anonymized starter dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. Total Scale of Inventory
total_rows = df.shape[0]
unique_pages = df['content_id'].nunique()
print(f"Total rows in starter slice: {total_rows}")
print(f"Unique content items: {unique_pages}")

# 2. Eligible Subset (Applying starter filters: impressions > 0 and age >= 90 days)
eligible_df = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)]
pct_eligible = (len(eligible_df) / total_rows) * 100
print(f"Eligible candidate pages for refresh analysis: {len(eligible_df)} ({pct_eligible:.2f}% of total)")

# 3. Magnitude of the Problem (Proportion of eligible pages currently declining)
trend_counts = eligible_df['trend_direction'].value_counts(normalize=True)
pct_down = trend_counts.get('down', 0) * 100
print(f"Proporção de páginas qualificadas em declínio estrutural ('down'): {pct_down:.2f}%")
```
